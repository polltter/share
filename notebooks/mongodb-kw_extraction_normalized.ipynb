```python
db.list_collection_names()
```

#### 1. Perform keyword extraction on a daily basis

We will start by running our kw_extraction.py script directly in order to have the **keyword extraction results normalized** for one day.

```python
# db['kw_freq_weight_test_norm'].drop()
```

```python
db.list_collection_names()
```

```python
kw_norm = db['kw_freq_weight_test_norm']
```

```python
for doc in kw_norm.find():
    print(doc)
```

```python
for doc in kw_norm.find():
    print(doc['extracted_at'])
```

```python
kw_norm.count_documents({})
```

We have extracted the most relevant keywords for our first 100 tweets (2022-09-29), and the results are now normalized (weight values range from 0 to 1).

We will now run the script again, changing the extraction date for the following days.

```python
kw_norm.count_documents({})
```

After running the script for all the available days, we get the results for all our tweets.
