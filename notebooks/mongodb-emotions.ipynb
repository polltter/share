#### 1. Perform emotion analysis on a daily basis

We will run our emotion_analysis.py script manually to simulate this analysis on a daily basis.

```python
db.list_collection_names()
```

We have created a collection named emotions_test.

```python
emotions_test = db['emotions_test']
```

```python
for doc in emotions_test.find():
    print(doc)
```

```python
emotions_test.count_documents({})
```

We will now run the script again, changing the extraction date for the following days.

```python
emotions_test.count_documents({})
```

After running the script for all the available days, we get the results for all our tweets.

```python
for doc in emotions_test.find():
    print(doc)
```
